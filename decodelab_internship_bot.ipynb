```python
import re
import datetime

tickets = []
feedback_log = []

BATCHES = {
    "1": "Batch 2025",
    "2": "Batch 2026",
    "3": "Batch 2027 (Upcoming)"
}

TRACKS = {
    "1": "Web Development",
    "2": "Mobile App Development",
    "3": "Graphic Designing",
    "4": "Digital Marketing",
    "5": "Machine Learning / AI",
    "6": "Database Systems",
}

FAQS = {
    "1": ("How long is the internship?",
          "The internship runs for 6-8 weeks depending on your selected track."),
    "2": ("Is the internship paid?",
          "This is an unpaid learning internship, but top performers get "
          "placement referrals for paid remote roles."),
    "3": ("How do I get my certificate?",
          "Certificates are issued within 7 working days after your final "
          "project is approved by your mentor."),
    "4": ("Can I switch tracks mid-internship?",
          "Track switches are allowed only within the first week -- contact "
          "human support to request one."),
    "5": ("What is the minimum attendance requirement?",
          "You need at least 80% attendance in mentor sessions to qualify "
          "for a certificate."),
}

def get_nonempty_input(prompt):
    while True:
        value = input(prompt).strip()
        if value:
            return value
        print("This field can't be empty. Please try again.")

def get_valid_email(prompt):
    pattern = r"^[\w\.-]+@[\w\.-]+\.\w+$"
    while True:
        value = input(prompt).strip()
        if re.match(pattern, value):
            return value
        print("That doesn't look like a valid email (e.g. name@example.com). Try again.")

def get_valid_phone(prompt):
    while True:
        value = input(prompt).strip()
        digits = value.replace("+", "").replace("-", "").replace(" ", "")
        if digits.isdigit() and 7 <= len(digits) <= 15:
            return value
        print("Please enter a valid phone number (digits only, 7-15 digits).")

def get_menu_choice(prompt, valid_options):
    while True:
        value = input(prompt).strip()
        if value in valid_options:
            return value
        print(f"Invalid option. Please choose from: {', '.join(valid_options)}")

def get_int_input(prompt, min_val=0, max_val=100):
    while True:
        value = input(prompt).strip()
        try:
            num = int(value)
            if min_val <= num <= max_val:
                return num
            print(f"Please enter a number between {min_val} and {max_val}.")
        except ValueError:
            print("Please enter a valid whole number.")

def print_header(title):
    print("\n" + "=" * 60)
    print(title.center(60))
    print("=" * 60)

def choose_from_dict(title, options_dict):
    print(f"\n{title}")
    for key, val in options_dict.items():
        label = val[0] if isinstance(val, tuple) else val
        print(f" {key}. {label}")
    return get_menu_choice("Enter your choice: ", list(options_dict.keys()))

def show_program_details():
    track_key = choose_from_dict("Available Internship Tracks:", TRACKS)
    track = TRACKS[track_key]
    print(f"\nYou selected: {track}")
    print("Duration : 6-8 weeks")
    print("Format : Remote, mentor-guided, project-based")
    print("Deliverable : One capstone project submitted for certification")

def check_application_status(intern_id):
    print(f"\nChecking status for Intern ID: {intern_id}...")
    if len(intern_id) % 2 == 0:
        print("Status: ACCEPTED - Welcome aboard! Check your email for onboarding steps.")
    else:
        print("Status: UNDER REVIEW - You'll hear back within 3-5 business days.")

def submit_task():
    print("\n--- Task / Project Submission ---")
    task_name = get_nonempty_input("Task or project title: ")
    file_name = get_nonempty_input("Attach filename (e.g. project.zip): ")
    notes = input("Any notes for your mentor (optional): ").strip()
    timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M")
    print(f"\nSubmission received at {timestamp}")
    print(f" Task : {task_name}")
    print(f" File : {file_name}")
    if notes:
        print(f" Notes : {notes}")
    print("Your mentor will review this within 2 working days.")

def mentor_and_batch_info():
    batch_key = choose_from_dict("Select your batch:", BATCHES)
    batch = BATCHES[batch_key]
    mentors = {
        "Batch 2025": "Mentor: Ayesha Khan | Sessions: Mon & Wed, 6-7 PM",
        "Batch 2026": "Mentor: Bilal Ahmed | Sessions: Tue & Thu, 7-8 PM",
        "Batch 2027 (Upcoming)": "Mentor: To be assigned | Starts next quarter",
    }
    print(f"\n{batch}")
    print(mentors[batch])

def certificate_info():
    print("\nCertificate Requirements:")
    print(" - Minimum 80% attendance")
    print(" - Final project approved by mentor")
    print(" - No pending fee/dues (if applicable)")
    eligible = get_menu_choice(
        "\nHave you submitted your final project? (yes/no): ",
        ["yes", "no"]
    )
    if eligible == "yes":
        print("Great! Your certificate will be processed within 7 working days.")
    else:
        print("Please submit your final project first to become eligible.")

def attendance_query():
    attended = get_int_input("How many sessions have you attended? ", 0, 100)
    total = get_int_input("How many sessions have been held so far? ", 1, 100)
    percent = (attended / total) * 100
    print(f"\nYour attendance: {percent:.1f}%")
    if percent >= 80:
        print("You meet the attendance requirement. Keep it up!")
    else:
        needed = max(0, int(0.8 * total) - attended)
        print(f"You're below the 80% threshold. Try to attend upcoming sessions "
              f"consistently to catch up.")

def show_faq():
    while True:
        key = choose_from_dict("Frequently Asked Questions:",
                                {k: v[0] for k, v in FAQS.items()})
        question, answer = FAQS[key]
        print(f"\nQ: {question}\nA: {answer}")
        again = get_menu_choice("\nView another FAQ? (yes/no): ", ["yes", "no"])
        if again == "no":
            break

def contact_human_support(name, email):
    issue = get_nonempty_input("Please describe your issue: ")
    urgency = get_menu_choice("Urgency (low/medium/high): ", ["low", "medium", "high"])
    ticket_id = f"TCK{len(tickets) + 1001}"
    tickets.append({
        "id": ticket_id, "name": name, "email": email,
        "issue": issue, "urgency": urgency
    })
    print(f"\nTicket created: {ticket_id}")
    print("Our support team will reach out to your registered email shortly.")

def remote_job_opportunities():
    print("\nPost-Internship Remote Opportunities")
    print("1. Python Developer")
    print("2. Front-End Developer")
    print("3. Back-End Developer")
    print("4. Full Stack Developer")
    job = get_menu_choice("Select Job: ", ["1", "2", "3", "4"])
    exp = get_int_input("Experience (years): ", 0, 50)

    job_names = {
        "1": "Remote Python Developer",
        "2": "Remote Front-End Developer",
        "3": "Remote Back-End Developer",
        "4": "Remote Full Stack Developer",
    }
    min_exp = {"1": 2, "2": 1, "3": 2, "4": 3}

    if exp >= min_exp[job]:
        print(f"Eligible for {job_names[job]}")
        apply = get_menu_choice("Apply now? (yes/no): ", ["yes", "no"])
        if apply == "yes":
            resume = get_nonempty_input("Resume filename: ")
            portfolio = input("Portfolio/GitHub link (optional): ").strip()
            print(f"\nApplication submitted for {job_names[job]}")
            print(f" Resume : {resume}")
            if portfolio:
                print(f" Portfolio: {portfolio}")
    else:
        print(f"Not eligible yet -- {job_names[job]} requires {min_exp[job]}+ years experience.")
        print("Tip: Completing more internship projects can help build relevant experience.")

def leave_feedback(name):
    rating = get_int_input("Rate your internship experience (1-5): ", 1, 5)
    comments = input("Any comments? (optional): ").strip()
    feedback_log.append({"name": name, "rating": rating, "comments": comments})
    print("Thank you for your feedback!")

def main():
    print_header("Decode Lab Internship Support Chatbot")
    print("Helping interns with programs, tasks, mentors, and more.\n")

    name = get_nonempty_input("Enter your name: ")
    email = get_valid_email("Enter your email: ")
    phone = get_valid_phone("Enter your phone number: ")
    intern_id = get_nonempty_input("Enter your Intern ID (or type NEW if you don't have one): ")

    print(f"\nHello {name}! Welcome to Decode Lab Internship Support.")

    menu_text = """
1. Internship Program Details
2. Check Application Status
3. Submit Task / Project
4. Mentor & Batch Info
5. Certificate Information
6. Attendance Check
7. Frequently Asked Questions
8. Contact Human Support
9. Remote Job Opportunities (Post-Internship)
10. Leave Feedback
11. Exit
"""

    while True:
        print(menu_text)
        choice = get_menu_choice(
            "Enter your choice (1-11): ",
            [str(i) for i in range(1, 12)]
        )

        if choice == "1":
            show_program_details()
        elif choice == "2":
            check_application_status(intern_id)
        elif choice == "3":
            submit_task()
        elif choice == "4":
            mentor_and_batch_info()
        elif choice == "5":
            certificate_info()
        elif choice == "6":


            attendance_query()
        elif choice == "7":
            show_faq()
        elif choice == "8":
            contact_human_support(name, email)
        elif choice == "9":
            remote_job_opportunities()
        elif choice == "10":
            leave_feedback(name)
        elif choice == "11":
            print(f"\nThank you for visiting Decode Lab, {name}. Good luck with your internship!")
            break

        cont = get_menu_choice("\nReturn to main menu? (yes/no): ", ["yes", "no"])
        if cont == "no":
            print(f"\nThank you for visiting Decode Lab, {name}. Good luck with your internship!")
            break

if __name__ == "__main__":
    main()
```
